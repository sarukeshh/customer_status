```python
# 📌 Step 2 — Load and Inspect Data
import pandas as pd
df = pd.read_csv("gold_customer_history.csv", parse_dates=["PurchaseDate"])
print("First 5 rows of dataset:")
display(df.head())

# 📌 Step 3 — Purchase Frequency
purchase_frequency = df.groupby("CustomerID")["PurchaseDate"].count().reset_index()
purchase_frequency.columns = ["CustomerID", "PurchaseCount"]

# 📌 Step 4 — Repeat Visitors
repeat_visits = purchase_frequency[purchase_frequency["PurchaseCount"] > 1]

# 📌 Step 5 — Avg Days Between Visits
df_sorted = df.sort_values(["CustomerID", "PurchaseDate"])
df_sorted["DaysSinceLastVisit"] = df_sorted.groupby("CustomerID")["PurchaseDate"].diff().dt.days
avg_return_time = df_sorted.groupby("CustomerID")["DaysSinceLastVisit"].mean().reset_index()
avg_return_time.columns = ["CustomerID", "AvgDaysBetweenVisits"]

# 📌 Step 6 — Items Bought
purchase_items = df.groupby("CustomerID")["ItemBought"].apply(list).reset_index()

# 📌 Step 7 — Merge into Customer Report
customer_report = purchase_frequency.merge(avg_return_time, on="CustomerID") \
                                    .merge(purchase_items, on="CustomerID")

print("\n📊 Customer Report:")
display(customer_report)

# 📌 Step 8 — Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Most Purchased Items
plt.figure(figsize=(8,5))
sns.countplot(x="ItemBought", data=df, order=df["ItemBought"].value_counts().index)
plt.title("Most Purchased Gold Items")
plt.xticks(rotation=45)
plt.show()

# Frequency Distribution
plt.figure(figsize=(8,5))
sns.histplot(purchase_frequency["PurchaseCount"], bins=range(1, purchase_frequency["PurchaseCount"].max()+2), kde=False)
plt.title("Distribution of Purchase Frequency")
plt.xlabel("Number of Purchases")
plt.ylabel("Number of Customers")
plt.show()

# Avg Days Between Visits
plt.figure(figsize=(8,5))
sns.histplot(avg_return_time["AvgDaysBetweenVisits"].dropna(), bins=10, kde=False)
plt.title("Average Days Between Visits")
plt.xlabel("Days")
plt.ylabel("Number of Customers")
plt.show()
```
